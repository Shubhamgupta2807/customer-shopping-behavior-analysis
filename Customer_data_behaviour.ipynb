```python
data["Review Rating"] = data.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
data.isnull().sum()
```

```python
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(" ","_")
data = data.rename(columns = {"purchase_amount_(usd)" : "purchase_amount"})
```

```python
data.columns
```

```python
#create a new age_group column.
labels = ["Young Adult", "Adult", "Middle_age", "Senior"]
data["age_group"] = pd.qcut(data["age"], q = 4, labels = labels)
```

```python
data[["customer_id","gender","age","age_group"]].head(5)
```

```python
#create new column purchase_frequency_days
frequently_mapping = {
    "Fortnightly" : 14,
    "Weekly" : 7,
    "Monthly" : 30,
    "Quarterly" : 90,
    "Bi-Weekly" : 14,
    "Annually" : 365,
    "Every 3 Months" : 90
} 

data["purchase_frequency_days"] = data["frequency_of_purchases"].map(frequently_mapping)
```

```python
data[["customer_id","purchase_frequency_days","frequency_of_purchases"]].head(5)
```
